```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import KFold
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(r'.\data mining for finance\WB.xls')
```

```python
df.plot()
```

```python
df.head()
```

```python
kf = KFold(df.shape[0],4)
```

```python
i=0
fig = plt.figure(figsize=(5,5))
for train,test in kf:
    ax = plt.subplot(2,2,i+1)
    ax.scatter(df.P[train],df.D[train],color='b')
    ax.scatter(df.P[test],df.D[test],color='r')
    i+=1
```
